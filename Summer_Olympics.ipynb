# **Summer Olympics Data Analysis Assignment**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tabulate import tabulate

df = pd.read_csv("summer.csv")
```

```python
df
```


**1. In how many cities Summer Olympics is held so far?**

```python
data = []

print("Summer olympics has been held in",len(df["City"].unique()),"different cities so far.")
print("\n\n")
for city in df["City"].unique():
  data.append([city , len(df[df['City']  == city])])

city_df = pd.DataFrame(data,columns = ['City' , 'Number of times']).sort_values(by='City', ascending=True)
print(city_df.set_index("City"))
print("\n\n")
plt.pie(city_df['Number of times'], labels = city_df['City'], radius = 2.5 , autopct = "%0.1f%%", pctdistance = 0.8)
plt.show()
```

**2. Which sport is having most number of Gold Medals so far? (Top 5)**

```python
data2 = []

for sport in df["Sport"].unique():
  sport_df = df[df["Sport"] == sport]
  data2.append([sport , len(sport_df[sport_df["Medal"] == "Gold"])])

gold_df = pd.DataFrame(data2,columns = ['Sport' , 'Number of gold medals']).sort_values(by='Number of gold medals', ascending=False).head()
print(gold_df.set_index("Sport"))
print("\n\n")
gold_df.head().plot(x = 'Sport', y = 'Number of gold medals', kind = 'bar', color = 'green', figsize = (5,3))
```

**3. Which sport is having most number of medals so far? (Top 5)**

```python
data3 = []

for sport in df["Sport"].unique():
  sport_df = df[df["Sport"] == sport]
  data3.append([sport , len(sport_df["Medal"])])

medal_df = pd.DataFrame(data3,columns = ['Sport' , 'Number of medals']).sort_values(by='Number of medals', ascending=False).head()
print(medal_df.set_index("Sport"))
print("\n\n")
medal_df.head().plot(x = 'Sport', y = 'Number of medals', kind = 'bar', color = 'violet', figsize = (5,3))
```

**4. Which player has won most number of medals? (Top 5)**

```python
data4 = []

for athlete in df["Athlete"].unique():
  athlete_df = df[df["Athlete"] == athlete]
  data4.append([athlete , len(athlete_df["Medal"])])

medal_df = pd.DataFrame(data4,columns = ['Athlete' , 'Number of medals']).sort_values(by='Number of medals', ascending=False).head()
print(medal_df.set_index("Athlete"))
print("\n\n")
medal_df.head().plot(x = 'Athlete', y = 'Number of medals', kind = 'bar', color = 'orange', figsize = (5,3))
```

**5. Which player has won most number Gold Medals of medals? (Top 5)**

```python
data5 = []

for athlete in df["Athlete"].unique():
  athlete_df = df[df["Athlete"] == athlete]
  data5.append([athlete , len(athlete_df[athlete_df["Medal"] == "Gold"])])

gold_df = pd.DataFrame(data5,columns = ['Athlete' , 'Number of gold medals']).sort_values(by='Number of gold medals', ascending=False).head()
print(gold_df.set_index("Athlete"))
print("\n\n")
gold_df.head().plot(x = 'Athlete', y = 'Number of gold medals', kind = 'bar', color = 'blue', figsize = (5,3))
```

**6. In which year India won first Gold Medal in Summer Olympics?**

```python
ind_df = df[df["Country"] == "IND"]
ind_df = ind_df[ind_df["Medal"] == "Gold"].min()["Year"]

print("India won gold for the first time in the year",ind_df)
```

**7. Which event is most popular in terms on number of players? (Top 5)**

```python
data7 = []

for event in df["Event"].unique():
  event_df = df[df["Event"] == event]
  data7.append([event , len(event_df["Athlete"].unique())])

popular_df = pd.DataFrame(data7,columns = ['Event' , 'Number of players']).sort_values(by='Number of players', ascending=False).head()
print(popular_df.set_index('Event'))
print("\n\n")

popular_df.plot(x = 'Event', y = 'Number of players', kind = 'bar', color = 'brown', figsize = (5,4))
```

**8. Which sport is having most female Gold Medalists? (Top 5)**

```python
data8 = []

women = df[df["Gender"] == "Women"]
gold = women[women["Medal"] == "Gold"]

for sport in gold["Sport"].unique():
  data8.append([sport , len(gold[gold["Sport"] == sport])])

women_df = pd.DataFrame(data8,columns = ['Sport' , 'Number of female gold medalists']).sort_values(by='Number of female gold medalists', ascending=False).head()
print(women_df.set_index("Sport"))
print("\n\n")
women_df.head().plot(x = 'Sport', y = 'Number of female gold medalists', kind = 'bar', color = 'red', figsize = (5,3))
```
